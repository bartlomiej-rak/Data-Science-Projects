# 2. Visual analysis

### Let's check the survival rate with respect to Pclass, Sex, SibSp, Parch and Embarked

```python
def surv_rate(df_train, feature_names, target_col):
    
    for feature in feature_names:

        rate = df_train.groupby(feature)[target_col].value_counts(normalize=True).unstack()
        
        fig, ax = plt.subplots()
        
        rate.plot(kind='bar', stacked=True, ax=ax)
        ax.set_xlabel(feature)
        ax.set_ylabel('Survival Rate')
        ax.set_title('Survival Rate with respect to ' + feature)
        ax.set_xticklabels(rate.index.tolist(), rotation=0)
```

```python
feature_names = ['Pclass', 'Sex', 'SibSp', 'Parch', 'Embarked']
target_col = 'Survived'

surv_rate(df_train, feature_names, target_col)
```

The following conclusions can be drawn:

1) It appears that the lower number of passenger class the higher the rate of survival.

2) It is apparent that women had much higher chances of survival (75% for women vs 20% for men)

3) In general, the lower the number of SibSp, the higher the rate of survival.

4) Similarly, smaller values for Parch [0:3] also result in significantly higher survival rate than for values [4:6].

5) It can be observed that there is a slight difference in survival rate between different ports of embarkment.

# 3. Data cleaning

```python
def clean_data(dataset):
    
    data = dataset
    
   # Fill missing Age using median of each sex belonging to each class.
    data['Age'] = data.groupby(['Pclass', 'Sex'])['Age'].apply(lambda x: x.fillna(x.median()))
    
    # drop PassengerId, Name, Cabin and Ticket columns
    data.drop(['PassengerId', 'Name', 'Cabin', 'Ticket'], axis=1, inplace=True)

    # drop samples where Embarked is NaN
    data.dropna(subset=['Embarked'], inplace=True)
    
    # Fill one missing Fare value in test dataset
    data['Fare'] = data['Fare'].fillna(data['Fare'].median())
    
    # reset index
    data.reset_index(inplace=True, drop=True)
    
    # Let's map the Sex feature with 0 for male and 1 for female
    data['Sex'] = data['Sex'].map({'male': 0, 'female': 1})
    
    #Do one-hot encoding for Embarked
    data = pd.get_dummies(data)
    
    return data
```

```python
# Check the data before cleaning
df_train.head(25)
```

```python
df_train_cleaned = clean_data(df_train)
```

```python
# Check the data after cleaning
df_train_cleaned.head(25)
```

# 4. Feature Importance Assessment using Random Forests

```python
X = df_train_cleaned.drop('Survived', axis=1).values
y = df_train_cleaned['Survived'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1, stratify=y)

feature_labels = df_train_cleaned.columns[1:]

forest = RandomForestClassifier(n_estimators = 800, random_state=1)

forest.fit(X_train, y_train)

importances = forest.feature_importances_

indices = np.argsort(importances)[::-1]

feature_labels_sorted = []

for f in range(X_train.shape[1]):
    print("%2d) %-*s %f" % (f + 1, 30,
                            feature_labels[indices[f]],
                            importances[indices[f]]))
    feature_labels_sorted.append(feature_labels[indices[f]])
          

plt.bar(range(X_train.shape[1]), importances[indices], align='center')
plt.xticks(range(X_train.shape[1]), feature_labels_sorted, rotation=90)

plt.title('Feature Importance')
plt.xlim([-1, X_train.shape[1]])
plt.tight_layout()
plt.show()
```

# 4. Model Building

```python
# Standardize the data

stdsc = StandardScaler()
X_train_std = stdsc.fit_transform(X_train)
X_test_std = stdsc.transform(X_test)
```

```python
def fit_and_predict(model):
    
    model.fit(X_train_std, y_train)
    y_pred = model.predict(X_test_std)
    
    return accuracy_score(y_test, y_pred)  
```

```python
model_1 = LogisticRegression(solver='liblinear', random_state=1)
model_2 = SVC()
model_3 = DecisionTreeClassifier()
model_4 = RandomForestClassifier()
model_5 = KNeighborsClassifier()

models = [model_1, model_2, model_3, model_4, model_5]

for idx, model in enumerate(models):
    print("\nModel", idx+1,":", model)
    print("ACC:", fit_and_predict(model))
```

```python
# Get the best estimator
model_4_opt = gs.best_estimator_
model_4_opt.fit(X_train, y_train)
print('Test accuract: %.3f' % model_4_opt.score(X_test, y_test))
```
